```python
df_list=list(df[['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER',
       'SALES', 'ORDERDATE', 'DAYS_SINCE_LASTORDER', 'MSRP']])
```

```python
plt.figure(figsize = (19,15))
for i in range(len(df_list)):
    plt.subplot(3, 6, i + 1)
    sns.boxplot(y = df[df_list[i]], data = df)
    plt.title('Boxplot of {}'.format(df_list[i]))
    plt.tight_layout()
```

```python
df1_list=list(df[['STATUS', 'PRODUCTLINE', 'COUNTRY', 'DEALSIZE' ]])
```

```python
# plt.figure(figsize = (20,8))
for i in range(len(df1_list)):
    plt.figure(figsize = (20,8))
    sns.countplot(data = df, x = df[df1_list[i]])
    plt.show() 
```
